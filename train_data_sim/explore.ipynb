```python
import cv2
import glob
from imgaug import augmenters as iaa
```

```python
import numpy as np
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('20171005_140823_state_0.jpg')
mimg = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(mimg)
```

```python
seq = iaa.Sequential([
    iaa.Fliplr(1.), # horizontally flip 50% of the images
    iaa.GaussianBlur(sigma=(0, 1.0)) # blur images with a sigma of 0 to 3.0
])
```

```python
aimg = seq.augment_image(img)
mimg = cv2.cvtColor(aimg, cv2.COLOR_BGR2RGB)
plt.imshow(mimg)
```

```python
unknown = [cv2.imread(f) for f in glob.glob('*_state_4.jpg')]
```

```python
img = unknown[np.random.randint(len(unknown))]
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
noise = (np.random.random(img.shape) * 256).astype(np.uint8)
plt.imshow(cv2.cvtColor(noise, cv2.COLOR_BGR2RGB))
```

```python
alpha = np.random.sample()
beta = 1. - alpha
c = cv2.addWeighted(img, alpha, noise, beta, 0.)
```

```python
plt.imshow(cv2.cvtColor(c, cv2.COLOR_BGR2RGB))
```

```python
seq = iaa.Sequential([
    iaa.Fliplr(1.), # horizontally flip 50% of the images
    iaa.AdditiveGaussianNoise(0, scale=(0,255), per_channel=0.8),
    iaa.GaussianBlur(sigma=(0, 1.0)) # blur images with a sigma of 0 to 3.0
])
```

```python

aimg = seq.augment_image(img)
mimg = cv2.cvtColor(aimg, cv2.COLOR_BGR2RGB)
plt.imshow(mimg)
```
